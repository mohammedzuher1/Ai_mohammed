```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("canada_per_capita_income.csv")
df
```

```python
plt.scatter(df.year, df['per capita income (US$)'], color='red', marker='+')
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Per Capita Income in Canada Over the Years')
plt.show()
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['year']], df['per capita income (US$)'])
```

```python
income_2020 = reg.predict([[2020]])
print(f"Predicted per capita income for Canada in 2020: {income_2020[0]}")
```

```python
print(f"Coefficient (slope): {reg.coef_[0]}")
print(f"Intercept: {reg.intercept_}")
```

```python
manual_income_2020 = 2020 * reg.coef_[0] + reg.intercept_
print(f"Manual prediction for 2020: {manual_income_2020}")
```
